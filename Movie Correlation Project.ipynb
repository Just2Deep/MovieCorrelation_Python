```python
# Import Libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize']=(12,8) #Adjusts the configuration of the plots we will create


# Read in the data

df = pd.read_csv('D:\python\Movie_Industry_Project\movies.csv')
```

```python
# looking at sample data
df.head()
```

```python
# Checking if there is any missing data

for col in df.columns:
    percent_missing = np.mean(df[col].isnull())
    print(f'{col} - {percent_missing}%')
```

```python
# dropping the missing data 

df = df.dropna()
```

```python
# datatypes of the columns
df.dtypes
```

```python
# removing the decimal zeros in budget and gross by converting it to an int64

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
df
```

```python
# Creating the correct year column from the released date 

df['YearCorrect'] = df['released'].str.extract(pat='([0-9]{4})').astype(int)
```

```python
# sorting the values in order of highest gross

df.sort_values(by='gross',inplace= False, ascending = False)
```

```python
# chaning the set option to pull up all the values from the datasets

pd.set_option('display.max_rows', 16)
```

```python
# Drop any duplicates if present

df.drop_duplicates()
```

```python
#ploting budget vs gross using scatter plot

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.ylabel('Gross Earnings in billions')
plt.xlabel('Budget for the film in 100 millions')

plt.show()
```

```python
#plotting the budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data =df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
# different correlations

df.corr(method = 'pearson') # diff methods pearson, kendall, spearman
```

```python
# high correlations between budget and gross

correlation_matrix = df.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation matrix for numeric features')

plt.ylabel('Movie features')
plt.xlabel('Movie features')

plt.show()
```

```python
# normalising/numerizing non numeric columns to find correlation between them

df_numerized = df

for col_name in df_numerized.columns:
    if df_numerized[col_name].dtype == 'object':
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
        
df_numerized
```

```python
df
```

```python
# correlation matrix for all values 

correlation_matrix = df_numerized.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation matrix for movie features')

plt.ylabel('Movie features')
plt.xlabel('Movie features')

plt.show()
```

```python
correlation_mat  = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
